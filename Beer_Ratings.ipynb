```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
print("There are {} rows and {} columns in the dataset".format(df.shape[0], df.shape[1]))
```

```python
df.info()
```

```python
style = df['Style'].value_counts().reset_index()
style
style.columns = ['Style', 'Counts']
style.head(15)
```

```python
fig = px.bar(
    style, 
    x = 'Style', 
    y = 'Counts', 
    color = 'Style', 
    title = 'Style Counts'
)
fig.update_layout(showlegend = False, width = 1000)
fig.show()
```

```python
brewery = df['Brewery'].value_counts().reset_index()
brewery.columns = ['Brewery', 'Beer Counts']
brewery
```

```python
brewery_fig = px.bar(
    brewery.head(15), 
    x = 'Brewery', 
    y = 'Beer Counts', 
    color = 'Brewery', 
    title = 'Brewery vs. Beer Counts'
)
brewery_fig.update_layout(showlegend = False, width = 1000)
brewery_fig.show()
```
